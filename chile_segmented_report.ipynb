# Segmented regression for assessing lockdown strategies for Chilean data

Segmented regression for assessing lockdown strategies for Chilean data using function 'segmented19' by Vito Muggeo (vito.muggeo@unipa.it)

Reference: Muggeo VMR, Sottile G, Porcu M. (2020) Modelling COVID-19 outbreak:
segmented regression to assess lockdown effectiveness (on ResearchGate)

doi: 10.13140/RG.2.2.32798.28485

## Load data

```r
chile <- read.csv(url("https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto3/CasosTotalesCumulativo_T.csv"))

dbnames <- c("Fecha", "Arica.y.Parinacota", "Tarapaca", "Antofagasta", "Atacama", "Coquimbo", "Valparaiso", "Metropolitana",
  "O.Higgins", "Maule", "Nuble", "Biobio", "Araucania", "Los.Rios", "Los.Lagos", "Aysen", "Magallanes", "Total")
names(chile) <- dbnames
```

## Load libraries and sources

```r
## loading required package
library(segmented)

## reading sources
source("segmented/segmented19.R")
```

```r
now <- proc.time()  # for speed test
```

## Fitting model using all data

```r
y <- chile$Total
attr(y, "first.date") <- as.character(chile$Fecha[1])
oChile <- segmented19(y, npsi = 1:10, rule = ">=0")
```

## Fitting model for each country region

```r
y <- chile$Arica.y.Parinacota
attr(y, "first.date") <- as.character(chile$Fecha[1])
oArica <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Tarapaca
attr(y, "first.date") <- as.character(chile$Fecha[1])
oTarapaca <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Antofagasta
attr(y, "first.date") <- as.character(chile$Fecha[1])
oAntofagasta <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Atacama
attr(y, "first.date") <- as.character(chile$Fecha[1])
oAtacama <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Coquimbo
attr(y, "first.date") <- as.character(chile$Fecha[1])
oCoquimbo <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Valparaiso
attr(y, "first.date") <- as.character(chile$Fecha[1])
oValpo <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Metropolitana
attr(y, "first.date") <- as.character(chile$Fecha[1])
oStgo <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$O.Higgins
attr(y, "first.date") <- as.character(chile$Fecha[1])
oOHiggins <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Maule
attr(y, "first.date") <- as.character(chile$Fecha[1])
oMaule <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Nuble
attr(y, "first.date") <- as.character(chile$Fecha[1])
oNuble <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Biobio
attr(y, "first.date") <- as.character(chile$Fecha[1])
oBiobio <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Araucania
attr(y, "first.date") <- as.character(chile$Fecha[1])
oAraucania <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Los.Rios
attr(y, "first.date") <- as.character(chile$Fecha[1])
oLosRios <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Los.Lagos
attr(y, "first.date") <- as.character(chile$Fecha[1])
oLosLagos <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Aysen
attr(y, "first.date") <- as.character(chile$Fecha[1])
oAysen <- segmented19(y, npsi = 1:10, rule = ">=0")

y <- chile$Magallanes
attr(y, "first.date") <- as.character(chile$Fecha[1])
oMagallanes <- segmented19(y, npsi = 1:10, rule = ">=0")
```
